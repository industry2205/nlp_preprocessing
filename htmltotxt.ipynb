```python
a = list(open('위키주소_건설.txt','r'))
def web_crawling(a):
    w = open('Dictionary_English.txt','w')
    c = []
    for i in a:
        webpage = requests.get(i[:-1])
        soup = BeautifulSoup(webpage.content, "lxml")
        
        for i in range(len(soup.find_all('p'))):
    
            a = str(soup.find_all('p')[i].get_text())
            a = re.sub('[[0-9]]','',a,0)
            a = re.sub('[[]','',a,0).strip()
            a = a.replace('\n','')
            b = re.split(r"\.",a)
            c.append(b)
    for d in c:
        d = d[0] + '\n'
        w.write(d)
    w.close()
```

```python
web_crawling(a)
```
